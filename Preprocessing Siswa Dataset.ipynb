### TABEL 4.3

```python
df = pd.read_csv(r'datapelanggan.csv')
df.columns = ['no', 'nama','alamat', 'bulan', 'jum_barang', 'nama_barang', 'jum', 'jenis_barang']

for index, row in df.iterrows():
    df.loc[index, 'new_bln'] = datetime_object = datetime.strptime(row['bulan'], '%Y-%m-%d')
```

```python
sd = dt.datetime(2015,6,1)
df['hist']=sd - df['new_bln']
df['hist'].astype('timedelta64[D]')
df['hist']=df['hist'] / np.timedelta64(1, 'D')
```

```python
rfmTable = df.groupby('nama').agg({'hist': lambda x:x.min(),
                                        'nama': lambda x: len(x),
                                        'jum_barang': lambda x: x.sum()})

rfmTable.rename(columns={'hist': 'recency', 
                         'nama': 'frequency', 
                         'jum_barang': 'monetary_value'}, inplace=True)
```
