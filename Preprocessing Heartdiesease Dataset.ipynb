```python
df = pd.read_csv('heartdiesease.csv')
df.columns = ['Umur','JK','JenisNyeriDada','TekananDarah','Kolesterol','PuasaGulaDara','RestingECG','MaksDetakJantung','ExerciseAngina','Oldpeak','ST_Slope','SakitJantung']
df.head()
```

### Seleksi Data

```python
d = df.loc[:,['Umur','JK','TekananDarah','Kolesterol','MaksDetakJantung','SakitJantung']]
d.head()
```

### Tranformasi Data

```python
data_new=d
for colname in d.select_dtypes(["object", "category"]):
        data_new[colname], _ = d[colname].factorize()
```
