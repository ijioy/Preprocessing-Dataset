```python
import pandas as pd
```

```python
df = pd.read_csv("datasiswa.csv")
df
```

```python
df.dtypes
```

```python
def_status = 'SLTA'
df.Pend.fillna(def_status, inplace = True)
df
```

```python
dummies1 = pd.get_dummies(df.PekerjaanWali)
dummies1.columns = ['ABRI','PNS','Pedagang','Petani','Wiraswasta']
dummies1
```

```python
dummies2 = pd.get_dummies(df.Pend)
dummies2.columns = ['D III','SD','SLTA','SLTP','SPG']
dummies2
```

```python
hasil = pd.concat([df,dummies1,dummies2],axis='columns')
hasil
```

```python
for index, row in df.iterrows():
    df.loc[index, 'thn_lahir'] = int(row['Tgl_Lahir'][-2:]) + 1900
for index, row in df.iterrows():
    df.loc[index, 'umur'] = int(row['lulus'] - int(row['thn_lahir']))

df[['Tgl_Lahir', 'thn_lahir', 'lulus', 'umur']]
```
